5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

```python
np.random.seed(42)
size = 100 
x1 = np.random.normal(50, 10, size)
x2 = x1 * 0.8 + np.random.normal(0, 5, size)
x3 = x1 * 0.5 + x2 * 0.3 + np.random.normal(0, 5, size) 
x4 = np.random.normal(50, 10, size) 


correlated_df = pd.DataFrame({
    "Feature1": x1,
    "Feature2": x2,
    "Feature3": x3,
    "Feature4": x4
})


plt.figure(figsize=(8, 6))
correlation_matrix = correlated_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)
pivot_sales = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')


fig = px.imshow(pivot_sales, 
                title="Average Sales per Day across Months", 
                labels={'x': 'Month', 'y': 'Day', 'color': 'Sales'},
                color_continuous_scale='Viridis')


fig.show()
```

4. Using the given x and y data, generate a 3D surface plot to visualize the function z=sin(rootover x^2 + y^2)

```python
import plotly.graph_objects as go

# Generate the x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)


fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

fig.update_layout(
    title="3D Surface Plot of z = sin(√(x² + y²))",
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    ),
    autosize=True
)

fig.show()
```

1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
from bokeh.plotting import figure, show
import numpy as np

# Generate x and y values
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a new plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

# Add a line renderer with x and y values
p.line(x, y, legend_label="Sine Wave", line_width=2)

# Show the plot
show(p)
```

2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_file
import numpy as np
from bokeh.models import ColumnDataSource


np.random.seed(42)
x = np.random.rand(100) * 10  
y = np.random.rand(100) * 10  
sizes = np.random.randint(5, 20, 100)  
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], 100)  


source = ColumnDataSource(data=dict(x=x, y=y, size=sizes, color=colors))


p = figure(title="Scatter Plot with Random Sizes and Colors", x_axis_label='x', y_axis_label='y')


p.scatter(x='x', y='y', size='size', color='color', legend_field='color', source=source, fill_alpha=0.6)


show(p)
```

3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.palettes import Spectral11


fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]


output_notebook()


p = figure(x_range=fruits, title="Fruit Counts", toolbar_location=None, tools="")


p.vbar(x=fruits, top=counts, width=0.9, color=Spectral11[:4], legend_field="Fruit")


p.xaxis.axis_label = "Fruit"
p.yaxis.axis_label = "Count"
p.title.text_font_size = '16pt'
p.legend.title = "Fruits"
p.legend.location = "top_left"


show(p)
```

4. Create a Bokeh histogram to visualize the distribution of the given data.

```python
from bokeh.plotting import figure, show
import numpy as np


data_hist = np.random.randn(1000)


hist, edges = np.histogram(data_hist, bins=30)


p = figure(title="Histogram of Generated Data", 
           x_axis_label='Value', 
           y_axis_label='Frequency', 
           tools="pan, box_zoom, reset")


p.quad(bottom=0, top=hist, left=edges[:-1], right=edges[1:], 
       fill_color="skyblue", line_color="white", alpha=0.7)


show(p)
```
